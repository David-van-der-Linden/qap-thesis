```python
import pandas as pd
import os
import sys
import matplotlib.pyplot as plt
import numpy as np


# make possible to import from src
notebook_dir = os.getcwd()
src_path = os.path.abspath(os.path.join(notebook_dir, '..', '..', 'src'))
if src_path not in sys.path:
    sys.path.append(src_path)

from my_secrets import my_path
```

```python
# * specify input
df = pd.read_pickle(my_path + 'visualizations/data_frames/prepared_exp_1.pkl')

# get instance names that have at least two distinct solving_technique entries
instances_with_both = df.groupby('instance_name')['solving_technique'].nunique()
instances_with_both = set(instances_with_both[instances_with_both >= 2].index)
print('len(instances_with_both):', len(instances_with_both))
# df filtered
dff = df[df['instance_name'].isin(instances_with_both)]
# df filtered main
dffm = dff[dff['experiment_name'] == 'main_experiment_1']
# df filtered variance
dffv = dff[dff['experiment_name'] == 'variance_test_experiment_1']
dffm.head()
```

```python
# only allow main experiment, where both kbl and dp have model status 2
dff1 = dffm.copy()
dff1 = dff1[(dff1['model.Status'] == 2)]
instances_with_both_f1 = dff1.groupby('instance_name')['solving_technique'].nunique()
instances_with_both_f1 = instances_with_both_f1[instances_with_both_f1 >= 2].index
dff1 = dff1[dff1['instance_name'].isin(instances_with_both_f1)]

df_pivot = dff1.pivot(index='instance_name', columns='solving_technique', values=['Runtime', 'total_time_in_user_cb', 'total_time_outside_user_cb'])

fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = np.arange(len(df_pivot))

# stacked bars for dp
bars1 = ax.bar(index, df_pivot[('total_time_outside_user_cb', 'dp')], bar_width, label='Time Outside CB (DP)', color='skyblue')
bars2 = ax.bar(index, df_pivot[('total_time_in_user_cb', 'dp')], bar_width, bottom=df_pivot[('total_time_outside_user_cb', 'dp')], label='Time Inside CB (DP)', color='orange')

# bar for kbl
bars3 = ax.bar(index + bar_width, df_pivot[('Runtime', 'kbl')], bar_width, label='Runtime (KBL)', color='gray')

ax.set_yscale('log')

ax.set_xlabel('Instance Name')
ax.set_ylabel('Time (s)')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(df_pivot.index, rotation=45)
ax.legend()

plt.tight_layout()
plt.savefig(my_path + 'visualizations/figures/stacked_time_comparison.png', format='png')
ax.set_title('Runtime of KBL vs stacked time Outside CB and time inside callback for DP, for instances that both methods solved till optimality')
plt.show()
```

```python
def get_instance_set(df_in: pd.DataFrame, status_code: int, solving_technique: str):
    """solving_technique can either be 'dp', 'kbl', or 'rlt'"""
    ldf = df_in.copy()
    ldf = ldf[ldf['model.Status'] == status_code]
    ldf = ldf[ldf['solving_technique'] == solving_technique]
    instance_list = ldf['instance_name'].to_list()
    instance_set = set(instance_list)
    if len(instance_list) != len(instance_set):
        raise Exception('expected one row per solving_technique and instance combination, got multiple')
    return instance_set


def get_instance_set_not(df_in: pd.DataFrame, not_status_codes: list[int], solving_technique: str):
    """solving_technique can either be 'dp', 'kbl', or 'rlt'"""
    ldf = df_in.copy()
    for status_code in not_status_codes:
        ldf = ldf[ldf['model.Status'] != status_code]
    ldf = ldf[ldf['solving_technique'] == solving_technique]
    instance_list = ldf['instance_name'].to_list()
    instance_set = set(instance_list)
    if len(instance_list) != len(instance_set):
        raise Exception('expected one row per solving_technique and instance combination, got multiple')
    return instance_set


dff2 = dffm.copy()
instances_st_2_dp = get_instance_set(dff2, 2, 'dp')
instances_st_2_kbl = get_instance_set(dff2, 2, 'kbl')
instances_st_9_dp = get_instance_set(dff2, 9, 'dp')
instances_st_9_kbl = get_instance_set(dff2, 9, 'kbl')
instances_st_17_dp = get_instance_set(dff2, 17, 'dp')
instances_st_17_kbl = get_instance_set(dff2, 17, 'kbl')
instances_rest_dp = get_instance_set_not(dff2, (2, 9, 17), 'dp')
instances_rest_kbl = get_instance_set_not(dff2, (2, 9, 17), 'kbl')

# * debug prints
print(f'\ninstances_st_2_dp, len:{len(instances_st_2_dp)}\n', instances_st_2_dp)
print(f'\ninstances_st_2_kbl, len:{len(instances_st_2_kbl)}\n', instances_st_2_kbl)
print(f'\ninstances_st_9_dp, len:{len(instances_st_9_dp)}\n', instances_st_9_dp)
print(f'\ninstances_st_9_kbl, len:{len(instances_st_9_kbl)}\n', instances_st_9_kbl)
print(f'\ninstances_st_17_dp, len:{len(instances_st_17_dp)}\n', instances_st_17_dp)
print(f'\ninstances_st_17_kbl, len:{len(instances_st_17_kbl)}\n', instances_st_17_kbl)
print(f'\ninstances_rest_dp, len:{len(instances_rest_dp)}\n', instances_rest_dp)
print(f'\ninstances_rest_kbl, len:{len(instances_rest_kbl)}\n', instances_rest_kbl)

if len(instances_rest_dp) != 0:
    raise Exception('statuscodes are unaccounted for in dp')
if len(instances_rest_kbl) != 0:
    raise Exception('statuscodes are unaccounted for in kbl')

if len(instances_st_2_dp) + len(instances_st_9_dp) + len(instances_st_17_dp) !=\
    len(instances_st_2_kbl) + len(instances_st_9_kbl) + len(instances_st_17_kbl):
    raise Exception('total kbl and dp does not match')

num_both = 0
num_neither = 0
num_only_dp = 0
num_only_kbl = 0

for inst in dff2['instance_name'].to_list():
    if inst in instances_st_2_dp and inst in instances_st_2_kbl:
        num_both += 1
    elif inst not in instances_st_2_dp and inst not in instances_st_2_kbl:
        num_neither += 1
    elif inst in instances_st_2_dp and inst not in instances_st_2_kbl:
        num_only_dp += 1
    elif inst not in instances_st_2_dp and inst in instances_st_2_kbl:
        num_only_kbl += 1
    else:
        raise Exception('unexpected instance is in behavior')
```

```python
values = [num_both, num_neither, num_only_dp, num_only_kbl]
total = sum(values)
print('total/2:', total/2)

labels = ['Both Methods', 'Neither Method', 'Only DP', 'Only KBL']

plt.bar(labels, values, color=['skyblue', 'orange', 'black', 'gray']) # '#33cc33'

# plt.xlabel('Categories')
plt.ylabel('Number of Instances')

for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha='center', va='bottom')

plt.savefig(my_path + 'visualizations/figures/optimality_with.png', format='png')
plt.title('Solved till optimality with')
plt.show()
```

```python
agg_data = dff2.groupby(['model.Status', 'solving_technique']).size().unstack(fill_value=0)

ax = agg_data.plot(kind='bar', figsize=(5, 4), color=['skyblue', 'gray'])

label_mapping = {
    'dp': 'DP',
    'kbl': 'KBL'
}

status_code_mapping = {
    2: 'Optimal',
    9: 'Timeout',
    17: 'MemLimit'
}

handles, labels = ax.get_legend_handles_labels()
new_labels = [label_mapping[label] for label in labels]
ax.legend(handles, new_labels, title='Solving Technique')

xticks_labels = [status_code_mapping.get(x, x) for x in agg_data.index]
ax.set_xticklabels(xticks_labels)

plt.xlabel('Model Status')
plt.ylabel('Number of Instances')
plt.xticks(rotation=0)
plt.savefig(my_path + 'visualizations/figures/num_instance_per_status.png', format='png')
plt.title('Number of Instances by Status Code and Solving Technique')
plt.show()
```
